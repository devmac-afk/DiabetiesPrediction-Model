```python
import pandas as pd
import numpy as np
```

```python
diabetes = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
diabetes
```

```python
diabetes.info()
```

```python
diabetes = diabetes.drop(['Fruits','Veggies','HvyAlcoholConsump','AnyHealthcare','NoDocbcCost','GenHlth','MentHlth','PhysHlth','DiffWalk','Sex','Education','Income'],axis=1)
```

```python
diabetes.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
ScaleMinMax = MinMaxScaler()
cols_to_scale = ['Diabetes_binary','HighBP','HighChol','CholCheck','BMI',
                 'Smoker','Stroke','HeartDiseaseorAttack','PhysActivity','Age']
diabetes = ScaleMinMax.fit_transform(diabetes)
```

```python
diabetes = pd.DataFrame(diabetes, columns=cols_to_scale)
diabetes
```

```python
X = diabetes.drop(columns=['Diabetes_binary','PhysActivity'],axis=1)
y = diabetes['Diabetes_binary']
```

```python
diabetes.to_csv("Diabetes.csv")
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_resampled
```

```python
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
train_pool = Pool(X_train, y_train)
test_pool = Pool(X_test, y_test)

import time
start = time.time()
# Train model
model = CatBoostClassifier()
model.fit(train_pool, eval_set=test_pool)

# Predict
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score

end = time.time()
diff = end - start
print("Time Taken: ",diff)
```

```python
from sklearn.metrics import classification_report
```

```python
model.get_feature_importance(prettified=True)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt


# Get feature importance
importances = model.get_feature_importance(train_pool)
feature_names = X_train.columns

# Create a DataFrame for easier viewing
feat_imp = pd.DataFrame({"Feature": feature_names, "Importance": importances})
feat_imp = feat_imp.sort_values(by="Importance", ascending=False)

print(feat_imp)

# Plot feature importance
plt.figure(figsize=(10,6))
plt.barh(feat_imp["Feature"], feat_imp["Importance"], color="skyblue")
plt.xlabel("Importance Score")
plt.title("CatBoost Feature Importance for Diabetes Prediction")
plt.gca().invert_yaxis()
plt.show()
```
